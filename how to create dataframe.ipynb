# How to combine two dataframes


```python
import pandas as pd

# Student1 DataFrame
student1 = pd.DataFrame({
    'rollno': [1, 2, 3, 4, 5],
    'name': ['ahsan', 'burhan', 'ahmad', 'abdullah', 'aman'],
    'age': [24, 24, 21, 12, 17]
})

print(student1.head(2))

# Student2 DataFrame
student2 = pd.DataFrame({
    'rollno': [9, 10, 11],
    'name': ['aman', 'javaid', 'majid'],
    'age': [32, 21, 34]
})

print(student2.head(2))
```

```python
#combining vertically

student_combined=pd.concat([student1,student2], axis=0)
student_combined
```

```python
## combining horixantally

student_combined=pd.concat([student1,student2], axis=1)
student_combined
```
